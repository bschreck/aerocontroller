```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
import time

def pltsin(ax, time, lines):
    if ax.lines:
        for i, line in enumerate(ax.lines):
            line.set_xdata(time)
            line.set_ydata(lines[i][1])
    else:
        for color, line in lines:
            ax.plot(time, line, color)
    fig.canvas.draw()

fig,axes = plt.subplots(2,1)
axes[0].set_xlabel('time')
axes[1].set_xlabel('time')
axes[0].set_ylabel('temp')
axes[1].set_ylabel('power')

len_time_to_plot = 30
axes[0].set_xlim(0, len_time_to_plot)
axes[1].set_xlim(0, len_time_to_plot)
while True:
    try:
        df = pd.read_csv(os.path.expanduser('~/kojifier/water_boiler.csv'))
    except pd.errors.EmptyDataError:
        time.sleep(0.1)
        continue
    times = df['times']
    powers = df['powers']
    temps = df['temps']
    setpoints = df['setpoints']
    if times.iloc[-1] - times.iloc[0] > len_time_to_plot:
        times_to_plot = times > times.iloc[-1] - len_time_to_plot
        times = times[times_to_plot]
        start_i = times.index[0]
        powers = powers[start_i:]
        temps = temps[start_i:]
        setpoints = setpoints[start_i:]
    
    low_diff, high_diff = (temps - setpoints).min() * 2, (temps - setpoints).max() * 2
    low = min(temps.min(), setpoints.min())
    high = max(temps.max(), setpoints.max())
    axes[0].set_ylim(low, high)
    axes[1].set_ylim(0, max(powers))
    
    pltsin(axes[0], np.linspace(0,30,len(times)), [('r', temps.values)])
    pltsin(axes[1], np.linspace(0,30,len(times)), [('b', powers.values)])
    axes[0].set_title(f'Setpoint = {setpoints.iloc[-1]}')
    fig.canvas.draw()
    time.sleep(0.2)
```
